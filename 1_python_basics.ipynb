### Conditions
- Operates by boolean
- `if condition == True`:
    - `print('yes')`
- `else`:
    - `print('no')`
 

```python
color = 'red'

if color == 'blue':
    print('yes')
else:
    print('no')
```

```python
color = 'yellow'

if color == 'blue':
    print('yes')
elif color == 'red':
    print('red')
else:
    print('no')
```

```python
"""
Try returning:
if age is between 15 - 17, print middle school student
if age is between 18 - 20, print high school student
if age is over 20, print adult
else, print else
"""

age = 18
if 15 <= age <= 17:
    print("middle school student")
elif 18 <= age <= 20:
    print('high school student')
elif 20 < age:
    print('adult')
else: 
    print('else')
```

### For Loops
- For iterating over a series of values and repeating certain calculation
- `for i in range(start_num, end_num, steps)`:
    - `Execute a function`
- by default, `for i in range(0, end_num, 1)`

```python
for i in range(10):
    print(i)
```

```python
for i in range(2, 20, 3):
    print(i)
```

```python
# Nested for loop
shirt = ['blue', 'purple', 'red']
pants = ['green', 'black', 'white']

for i in shirt:
    for j in pants:
        print(i, j)
```

```python
"""
Try returning:
1 ~ 9단 구구단
1 X 1
1 X 2
1 X 3
1 X 4
1 X 5
...
9 X 9
"""
for i in range(1, 9):
    for j in range(1, 9):
        print(i, 'X', j)
```

Adding elements to an empty list

`break`: stop iterating and exit out of for loop \
`continue`: loop back to the beginning of the loop

```python
lst = ['hi', 'hello', 'bye']
for i in lst:
    if i == 'hello':
        continue
    print(i)
```

### While Loop
- For repeating a function until it reaches certain condition
- `while condition:` 
    - `execute a function`

```python
count = 10

while count > 0:
    print(count)
    count = count - 1
```

```python
# watch out for infinite loop!
count = 0
while True:
    print("Infinite loop")
    count += 1
    if count == 1000:
        break
    
```

```python
# break in while loop
status = True
result = 0

while status:
    result += 1
    if result > 5:
        status = False

print(result)
```

```python
"""
Try implementing a coffee vending machine
Coffee is 300 won.

If a person pays 300 won, sell coffee
If a person pays less than 300 won, return the money and don't sell coffee
If a person pays more than 300 won, sell coffee and return change
If coffee is sold out, break while loop and stop selling

"""

coffee = 5
while True:
    money = int(input("돈을 넣어 주세요: "))
    if money == 300:
        print("커피를 줍니다.")
        coffee = coffee -1
    elif money > 300:
        print("거스름돈 {}를 주고 커피를 줍니다.".format(money - 300))
        coffee = coffee - 1
    else:
        print("돈을 다시 돌려주고 커피를 주지 않습니다.")
        print("남은 커피의 양은 {}개 입니다.".format(coffee))
    if coffee == 0:
        print("커피가 다 떨어졌습니다. 판매를 중지 합니다.")
        break
```
