## Dataframe - Aggregation, groupby

### 1. Sorting
- `pd.sort_values(param)`
    - `by`, `ascending`, `inplace`

```python
import pandas as pd
import numpy as np
```

```python
"""
Try this:
Import train.csv as a dataframe
"""
data = pd.read_csv('titanic/train.csv')
data
```

```python
# by default ascending = True
sorted_data = data.sort_values(by = ['PassengerId'])
sorted_data
```

```python
sorted_data = data.sort_values(by = ['Name', 'Age'], ascending = False)
sorted_data
```

```python
"""
Try this:
Sort by Descending values in 'Name' and 'Fare'
"""
sorted_data = data.sort_values(by = ['Fare', 'Name'], ascending = False)
sorted_data
```

### 2. Aggregation
- `min(), max(), sum(), count(), mean(), median()` function on the columns of dataframe

```python
# returns the minimum value of each columns
data.sum()
```

```python
# returns the count of every columns
data.count()
```

```python
data.info()
```

```python
# calculate only certain rows
data[['Age', 'Fare']].median()
```

### 3. Groupby
- `dataframe.groupby(by = COLUMN_NAME).function()`

```python
data
```

```python
# group by the values of column Pclass and count the values
data = data.groupby(by = 'Pclass')['Age'].mean()
data
```

```python
# get only certain columns from groupby
data_grouped = data.groupby(by = 'Pclass')[['Age', 'Fare']].mean()
data_grouped
```

```python
"""
Try this:
Get average of 'Age' and 'Survived' of different 'Sex'
"""
data_average = data.groupby(by = 'Sex')[['Age', 'Survived']].mean()
data_average
```

### 4. Defining Functions
- `def FUNC_NAME(param):`
    - `execution`

```python
def addtwo(num1, num2):
    return num1 + num2

addtwo(3, 2)
```

```python
def execute():
    print('Done')

execute()
```

```python
"""
Try this:
define a function 'calc' with parameters 'column'
which returns a mean of input column of a dataframe
"""
def calc(column):
    return data[column].mean()

calc('Age')
```
